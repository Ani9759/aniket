Langchain- LangChain is a framework for developing applications powered by language models. It enables applications that are:

            - Data-aware: connect a language model to other sources of data
            - Agentic: allow a language model to interact with its environment
LangChain facilitates the seamless integration of retrieval and augmentation steps, making it easier to work with LLMs and Python.

Large Language Model- It is a type of pretrained model i.e LLM is a model which is trained of many terabytes text file. It is used to extract the information from their training data. It is a technique by which we predict something from their trained data. Language Models (LMs) have revolutionized NLP with their ability to understand and generate human-like text. In retrieval augmented generation, LLMs play a pivotal role in retrieving and generating contextually relevant information.

Retrieving augmentation generation- It is advance version of Large language model. There are some drawbacks by using Large language model only i.e in Large language model the prediction is based on their traing data so if any update is come on that request so that the answer which large language give is wrong and second drawback to use large language model only is that there is no proof that this is a correct information whereas in Retrieving Augmentation Generation we use large language model as well as vector data base so if any update is come so we simply update our vector database and in Retrieving augmentaion there is a proof that this is a correct information.

To begin with, first define a Python dictionary, named, config to store all URLs for the data sources and list of queries to be asked to our Recipe Retriever System, as following:

```python
config = {
    "websites": [
        "https://www.101cookbooks.com/panade/",
        "https://www.101cookbooks.com/good-chana-masala-recipe/",
        "https://www.101cookbooks.com/tofu-scramble/",
    ],
    "questions": [
        "What are the top 5 ingredients in 'Chana Masala'?",
        "How to make a delicious 'Tofu Scramble'?",
        "Do you know how to make 'Paneer Tikka'?",
        "I have these items in my bucket: 'broccoli florets', 'cabbage', 'cauliflower florets', 'zucchini'. What's the dish that I can prepare?"
    ],
}
```

Now, import WebBaseLoader from langchain library to load the data from the list of URLs, as following:

```python
from langchain.document_loaders import WebBaseLoader

# Define the data source for the loader
loader = WebBaseLoader(config["websites"])

# Load the data from URLs
docs = loader.load()

# Print the length of docs read
print(len(docs))

# Print the metadata from first doc
print(docs[0].metadata)
```

Prepare the data source 
  -->split the loaded documents

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Function to split the list of documents loaded from URLs
def split_docs(documents, chunk_size=1000, chunk_overlap=10):
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
    docs = text_splitter.split_documents(documents)
    return docs

docs = split_docs(docs)
print(len(docs))
```

-->Index the data

```python
from langchain.embeddings import SentenceTransformerEmbeddings
embeddings = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
```

Store the indices in Chroma DB

```python
pip install chromadb
```

```python
from langchain.vectorstores import Chroma

# Store embeddings for the documents
db = Chroma.from_documents(docs, embeddings)
```

Use of Large language model to retrieve the relevant information

```python
from dotenv import load_dotenv
from langchain.chat_models import ChatOpenAI
from langchain.chains import RetrievalQA

# Load the environment variables using .env file
load_dotenv(override=True)

# Define LLM using OpenAI GPT 3.5 model
model_name = "gpt-3.5-turbo"
llm = ChatOpenAI(model_name=model_name,openai_api_key="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")

# Define the retrieval chain
retrieval_chain = RetrievalQA.from_chain_type(llm, chain_type="stuff", retriever=db.as_retriever())
```

Apply the LangChain for augmentation

```python
import time

for q in config["questions"]:
    # Fetch the relevant documents for query and answer using the context
    ans = retrieval_chain.run(q)
    print(ans)
    print("---")
    time.sleep(5)
```
